```python
%sql
CREATE CATALOG IF NOT EXISTS dbx_course; -- this is like a Database in Snowflake
USE CATALOG dbx_course;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS source;
CREATE SCHEMA IF NOT EXISTS target;
```

```python
%sql
CREATE VOLUME IF NOT EXISTS dbx_course.source.files
```

```python
# Check if data already exists before copying
try:
    files = dbutils.fs.ls("/Volumes/dbx_course/source/files/")
    if len(files) > 0:
        print("Source data already exists, skipping copy")
    else:
        raise Exception("Data files are note present")
except:
    print("Copying data from S3...")
    #spark.conf.set("fs.s3a.bucket.dbx-data-public.aws.credentials.provider", 
    #               "org.apache.hadoop.fs.s3a.AnonymousAWSCredentialsProvider")
    dbutils.fs.cp("s3a://dbx-data-public/v03/", "/Volumes/dbx_course/source/files/", recurse=True)
    print("Done!")
```
